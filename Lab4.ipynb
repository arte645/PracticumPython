```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
print("Задание 1:")
f = lambda x: math.sin(math.sqrt(2*x))
a = 0
b = 1
segments = 5

x = np.linspace(a, b, segments)
y = []
for i in range(len(x)):
    y.append(f(x[i]))

for i in range(segments):
    print(f"В точке {x[i]}, значение функции {round(y[i], 5)}.")
```

```python
plt.plot(x, y)

plt.title("Задание 2:", fontsize=14)
plt.xlabel("x", fontsize=14)
plt.ylabel("y", fontsize=14)

plt.legend(["y = sin(sqrt(2x))"])

plt.show()
```

```python
plt.title("Задание 3: Точечный график")
plt.scatter(x, y, marker="x", color = (0.1, 0.9, 0.9))

plt.xlabel("x")
plt.ylabel("y")


plt.grid(color='black', linestyle=':', linewidth=0.5)

plt.show()
```

```python
uniform_vec = np.random.uniform(0, 10, size=100).astype(int)
normal_vec = np.random.normal(0, 10, size=100).astype(int)

plt.hist(uniform_vec, bins=100, color="black")
plt.title("Задание 4: Равномерное распределение")
plt.show()
plt.hist(normal_vec, bins=99, color="grey")
plt.title("Нормальное распределние")
plt.show()
```

```python
uniform = np.random.uniform(1, 5, 50).astype(int)
res = np.unique(uniform, return_counts=True)

numbers = res[0]
sizes = res[1]

plt.pie(x = sizes, labels=numbers, counterclock=True, autopct='%1.0f%%', colors=("green", "grey", "brown", "blue"))
plt.title("Задание 5: круговая диаграмма")
plt.show()



plt.bar(numbers, sizes, align='center', color=("green", "grey", "brown", "blue"))
plt.xticks(numbers)
plt.ylabel('Кол-во')
plt.title('Столбчатую диаграмма')

plt.show()
```

```python
f = lambda x,y: (x + 3)**2 + (x - 3)**2

x = np.linspace(a, b, segments)
y = np.linspace(a, b, segments)
X, Y = np.meshgrid(x, y)

Z = f(X, Y)


axes = plt.figure().add_subplot(projection='3d')
axes.set_xlabel("x", fontdict={"size": 14})
axes.set_ylabel("y", fontdict={"size": 14})
axes.set_zlabel("z", fontdict={"size": 14})


axes.plot_surface(X, Y, Z, color="red")
axes.set_title("Задание 6:")

plt.show()
```

```python
def experiments(style,num):
    f = lambda x,y: (x + 3)**2 + (x - 3)**2

    x = np.linspace(a, b, segments)
    y = np.linspace(a, b, segments)
    X, Y = np.meshgrid(x, y)

    Z = f(X, Y)
    plt.style.use(style)

    fig = plt.figure(figsize=(5, 5), layout="constrained")
    fig.suptitle(f"Задание {num}:", fontsize="20")

    ax3 = fig.add_subplot(223)
    ax3.set_title("График 3")
    ax3.pie(x = sizes, labels=numbers, autopct='%1.0f%%', colors=("green", "grey", "brown", "blue"))

    ax4 = fig.add_subplot(224, projection="3d")
    ax4.plot_surface(X, Y, Z, color="red")
    ax4.set_title("График 4")

    f = lambda x: math.sin(math.sqrt(2*x))
    x = np.linspace(a, b, segments)
    y = []
    for i in range(len(x)):
        y.append(f(x[i]))

    ax1 = fig.add_subplot(221)
    ax1.set_title("График 1")
    ax1.set_facecolor("lightyellow")
    ax1.plot(x, y)

    ax2 = fig.add_subplot(222)
    ax2.set_title("График 2")
    ax2.scatter(x, y, marker="x", color = (0.1, 0.9, 0.9))

    plt.show()

experiments("default",7)
```

```python
experiments("classic",8.1)
experiments("dark_background",8.2)
experiments("seaborn-dark",8.3)
experiments("Solarize_Light2",8.4)
```
