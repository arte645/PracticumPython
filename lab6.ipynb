Удаляем дубликаты


```python
data=data.drop_duplicates()
```

Удаляем id:

```python
data=data.drop(['Customer ID'],axis=1)
```

Меняем название столбца:

```python
data.rename(columns={"Gender":"Sex"}, inplace=True)
```

```python
data
```

Гистограмма:

```python
n_bins = 20
plt.hist(data['Review Rating'], bins=n_bins)
plt.savefig('1.png')
```

```python
plt.boxplot(x=data['Age'])
plt.savefig('2.png')
```

Круговая диаграмма:

```python
fig, axs = plt.subplots(figsize=(25, 25))

count_numbers = np.bincount(data['Age'])

axs.pie(count_numbers[1:], autopct='%1.1f%%', startangle=140)
axs.set_title('Возраст',)
fig.savefig('3.png')
```

Тепловая карта:

```python
plt.figure(figsize = (12,15))
sns.heatmap(data.corr(),annot=True,cmap="coolwarm",fmt=".2f")
plt.savefig('4.png')
```

Диаграмма countplot:

```python
f = plt.figure(figsize=(20,10))
def mapper(value):
  if value > 60:
    return "Old" 
  elif value > 40: 
    return "Middle"
  else: return "Young"
data["Age_mark"] = data["Age"].map(mapper)
sns.countplot(x = data["Review Rating"], hue=data["Age_mark"], )
plt.savefig('5.png')
```
