```python
import pandas as pd
```

```python
print("Задание 1:")
df1 = pd.DataFrame({
    'Name' : ['Strom, Mrs. Wilhelm (Elna Matilda Persson)', 'Navratil, Mr. Michel ("Louis M Hoffman")', 'Minahan, Miss. Daisy E'],
    'Age' : [29, 36.5, 33],
    'Sex' : ['female', 'male', 'female']
})
df1
```

```python
print("Задание 2:")
df2 = pd.read_csv('titanic_csv.csv', sep=';')
df2.head(10)
```

```python
print("Задание 3:")
url = 'https://gist.githubusercontent.com/zaryanezrya/8b4ef51c707cb16d5e88a44dc00a1bb2/raw/41230f49c6268e072dbf102672f670be256922ab/gistfile1.txt'
df3 = pd.read_csv(url)
df3.head()
```

```python
print("Задание 4:")
renamed_columns = {}
for column in df2.columns:
    renamed_columns[column] = column.lower().strip()
df2 = df2.rename(columns = renamed_columns)
df2.head()
df23 = pd.concat([df2, df3])
df23.duplicated().sum()
df23.drop_duplicates(inplace=True)
df23.duplicated().sum()
```

```python
print("Задание 5:")
df23 = df23.set_index('passengerid').sort_index()
df23.head(10)
```

```python
print("Задание 6:")
df23.info()
```

```python
df23.describe()
```

```python
print("Задание 7:")
df23.loc[2,:]
df_copy=df23.copy()
df23.iloc[0,:], df23.loc[2,:]=df_copy.loc[2,:],df_copy.iloc[0,:]
df23.head(4)
```

```python
print("Задание 8:")

df23['sex'] = df23['sex'].map({'female': 'f','male' : 'm'})
df23['sex'].unique()
```

```python
print("Задание 9:")

group = df23.groupby('ticket')['name'].count().rename("value").reset_index()
group[group['value'] >= 6].sort_values(by = 'value', ascending = False)
```
